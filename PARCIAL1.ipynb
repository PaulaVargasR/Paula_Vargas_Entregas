```python
urbano=0;
rural=0;
mu=0;
fu=0;
mr=0;
fr=0;
for i in range(395):
    if student['address'][i]=='U':
        urbano=urbano+1;
        if(student['sex'][i])=='M':
           mu=mu+1;
        else:
           fu=fu+1;
        
    else:
        rural=rural+1;
        if(student['sex'][i])=='M':
           mr=mr+1;
        else:
           fr=fr+1;

print('La cantidad de hombres que viven en zona rural es de:',mr)
print('La cantidad de hombres que viven en zona urbana es de:',mu)
print('La cantidad de mujeres que viven en zona rural es de:',fr)
print('La cantidad de mujeres que viven en zona urbana es de:',fu)
```

```python
trans=0;
m=0;
f=0;

for i in range(395):
  if student['traveltime'][i]==4:
    trans=trans+1;
    if student['sex'][i]=='M':
      m=m+1;
    else:
      f=f+1;

print('Los estudiantes (masculinos) que se demoran más de una hora en transportarse son:',m)
print('Los estudiantes (femeninos) que se demoran más de una hora en transportarse son:',f)
```

```python
fallas=0;
m=0;
f=0;

for i in range(395):
  if student['failures'][i]>=5:
    fallas=fallas+1;
    if student['sex'][i]=='M':
      m=m+1;
    else:
      f=f+1;


print('Los estudiantes (masculinos) que tienen más de 5 fallas son:',m)
print('Los estudiantes (femeninos) que tienen más de 5 fallas son:',f)
```

```python
medu=0;
fedu=0;

for i in range(395):
  if student['Medu'][i]==4:
    medu=medu+1;
  if student['Fedu'][i]==4:
    fedu=fedu+1;

print('Los estudiantes que tienen padres con un grado de escolaridad superior a la educación secundaria son:',medu)
print('Los estudiantes que tienen madres con un grado de escolaridad superior a la educación secundaria son:',fedu)
```
